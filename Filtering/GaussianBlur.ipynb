```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from google.colab import drive
from google.colab.patches import cv2_imshow
drive.mount('/content/drive')
```

## 귀여운 리트리버 사진 불러오기

```python
img=cv.imread('drive/MyDrive/Colab Notebooks/Pattern Recognition/강아지2.PNG', 1) # 3ch
plt.figure(figsize=(5,5))
cv2_imshow(img)
```

## cv.GaussianBlur 적용 전, 후 비교

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
original_image=cv.cvtColor(img, cv.COLOR_BGR2RGB) # bgr -> rgb
plt.imshow(original_image)
plt.title('Original Image', size=15)
plt.subplot(1,2,2)
Gaussian_image=cv.GaussianBlur(img,(5,5), 0) # (5,5)의 kernelsize를 가지는 GaussianBlur
Gaussian_image=cv.cvtColor(Gaussian_image, cv.COLOR_BGR2RGB) # bgr -> rgb
plt.imshow(Gaussian_image) 
plt.title('After Gaussian', size=15)
```

## 티가 잘 안나 일부분 확대해서 살펴보았습니다.

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.imshow(original_image[200:500, 400:700])
plt.title('Original Image', size=15)
plt.subplot(1,2,2)
plt.imshow(Gaussian_image[200:500, 400:700]) 
plt.title('After Gaussian', size=15)
```

> 확실히 우측의 이미지가 블러처리된 것을 볼 수 있습니다.
